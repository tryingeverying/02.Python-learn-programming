### 拷贝页面

```python
import PyPDF2
pdf1file = open('PDF文档\meetingminutes.pdf','rb')
pdf2file = open('PDF文档\meetingminutes2.pdf','rb')
```

```python
pdf1reader = PyPDF2.PdfFileReader(pdf1file)
pdf2reader = PyPDF2.PdfFileReader(pdf2file)
```

```python
pdfwriter = PyPDF2.PdfFileWriter()
```

```python
# 遍历文档的每一页，将之赋予页对象
for pagenum in range(pdf1reader.numPages):
    pageobj = pdf1reader.getPage(pagenum)
    pdfwriter.add_page(pageobj)     # .add_page(pageobj) 在文档的末尾添加页面

for pagenum in range(pdf2reader.numPages):
    pageobj = pdf2reader.getPage(pagenum)
    pdfwriter.add_page(pageobj)
```

```python
pdfoutputfile  = open('PDF文档\combinedminutes.pdf','wb') 
```

```python
pdfwriter.write(pdfoutputfile)
pdfoutputfile.close()
pdf1file.close()
pdf2file.close()
```

### 旋转页面

```python
import PyPDF2
minutesfile =  open('PDF文档\meetingminutes.pdf','rb')
```

```python
pdfreader = PyPDF2.PdfFileReader(minutesfile)
```

```python
page = pdfreader.getPage(0)
```

```python
page.rotate_clockwise(90)       #.rotate_clockwise(x) 可对PDF文件的页面进行90 180 270 的旋转
```

```python
pdfwriter = PyPDF2.PdfFileWriter()
```

```python
pdfwriter.addPage(page)
resultpdffile = open(r'PDF文档\rotatepage.pdf','wb')
```

```python
pdfwriter.write(resultpdffile)
```

```python
resultpdffile.close()
minutesfile.close()
```

### 叠加页面

```python
import PyPDF2
minutefile =  open('PDF文档\meetingminutes.pdf','rb')
pdfreader = PyPDF2.PdfFileReader(minutefile)
minutefirstpage = pdfreader.getPage(0)
```

```python
pdfwatermarkreader = PyPDF2.PdfFileReader(open('PDF文档\watermark.pdf','rb'))
minutefirstpage.merge_page(pdfwatermarkreader.getPage(0))    #A.merge_page(B) 将A和B合并到一页 
pdfwriter.add_page(minutefirstpage)
```

```python
for pagenum in range(1,pdfreader.numPages):
    pageobj = pdfreader.getPage(pagenum)
    pdfwriter.addPage(pageobj)
resultpdffile = open('PDF文档\watermarkcover.pdf','wb')
pdfwriter.write(resultpdffile)
```

```python
minutesfile.close()
resultpdffile.close()
```

### 加密pdf

```python
import PyPDF2
pdffile =  open('PDF文档\meetingminutes.pdf','rb')
pdfreader = PyPDF2.PdfFileReader(pdffile)
pdfwriter = PyPDF2.PdfFileWriter()
```

```python
for pagenum in range(pdfreader.numPages):
    pdfwriter.addPage(pdfreader.getPage(pagenum))
```

```python
pdfwriter.encrypt('swordfish')     #.encrypt(x) 以x为密码对文档进行加密  
resultpdf = open('PDF文档\swordfish.pdf','wb')
pdfwriter.write(resultpdf)
resultpdf.close()
```
